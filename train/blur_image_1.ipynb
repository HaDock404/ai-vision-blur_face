```python
import cv2 #attention installer une version de numpy inférieur à 2.0.0 
# $ pip install "numpy<2.0.0"
import matplotlib.pyplot as plt
```

```python
# Charger l'image
image = cv2.imread('image.jpg')

# Définir la région de l'image à flouter
# x, y, largeur, hauteur (top-left corner, width, height)
x, y, w, h = 100, 100, 200, 200  # Change ces valeurs selon tes besoins

# Extraire la région d'intérêt (ROI)
roi = image[y:y+h, x:x+w]

# Appliquer un flou gaussien à la région
roi_floutee = cv2.GaussianBlur(roi, (51, 51), 0)

# Remplacer la région originale par la version floutée
image[y:y+h, x:x+w] = roi_floutee

# Enregistrer l'image résultante
cv2.imwrite('image_floutee.jpg', image)
```

```python
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```
